```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil import tz
```

```python
data_filepath = "C:/Users/glius/Google Drive/HOPE/data"
def read_data(option, data_filepath):
  raw_data = pd.DataFrame()
  patient_names = os.listdir(data_filepath)
  for patient_name in patient_names:
    data_folder = data_filepath + "/" + patient_name +  "/" + option
    if os.path.exists(data_folder):
      data_files = os.listdir(data_folder)
      for data_file in data_files: 
        dest_path = data_folder + "/" + data_file
        hour_data = pd.read_csv(dest_path)
        hour_data["patient"] = patient_name
        from_zone = tz.gettz('UTC')
        to_zone = tz.gettz('America/New_York')
        UTC = [datetime.strptime(i, '%Y-%m-%dT%H:%M:%S.%f') for i in hour_data['UTC time']]
        EST = [i.replace(tzinfo=from_zone).astimezone(to_zone) for i in UTC]
        hour_data["date"] = [i.date() for i in EST]
        raw_data = raw_data.append(hour_data)
  return raw_data
```

```python
raw_calls = read_data("calls",data_filepath)
raw_texts = read_data("texts",data_filepath)
```

```python
## Compute statistics from the call data
raw_calls['duration in seconds'] = raw_calls['duration in seconds'].fillna(0.0).astype(int)
raw_calls['duration in minutes'] = raw_calls['duration in seconds']/60

summary_calls = raw_calls.groupby(["patient","date","call type"]).agg(
    {
        'timestamp': "count",
        'hashed phone number': "nunique",
        'duration in minutes': ["sum","mean"]
    }
)
calls_stats = summary_calls.unstack()
calls_stats = calls_stats.fillna(0.0)
calls_stats['id'] = [list(calls_stats.index)[i][0] for i in range(calls_stats.shape[0])]
calls_stats['date'] = [list(calls_stats.index)[i][1] for i in range(calls_stats.shape[0])]
calls_stats.columns = ['num of in calls','num of missed calls','num of out calls',
                       'num of in phones','num of missed phones','num of out phones',
                       'total len of in calls','total len of missed calls','total len of out calls',
                       'mean len of in calls','mean len of missed calls','mean len of out calls',
                       'id','date']
new_calls = calls_stats
calls_stats = pd.DataFrame(columns = list(calls_stats.columns))
for i in calls_stats.columns:
  calls_stats[i] = list(new_calls[i])
del new_calls
calls_stats.head(10)
```

```python
## Compute statistics from the text data
raw_texts.loc[raw_texts["sent vs received"] == "sent MMS", "sent vs received"] = "sent SMS"
raw_texts.loc[raw_texts["sent vs received"] == "received MMS", "sent vs received"] = "received SMS"
raw_texts.loc[raw_texts["message length"] == "MMS", "message length"] = 0
raw_texts['message length'] = raw_texts['message length'].fillna(0.0).astype(int)

summary_texts = raw_texts.groupby(["patient","date","sent vs received"]).agg(
    {
        'timestamp': "count",
        'hashed phone number': "nunique",
        'message length': ["sum","mean"]
    }
)
texts_stats = summary_texts.unstack()
texts_stats = texts_stats.fillna(0.0)
texts_stats['id'] = [list(texts_stats.index)[i][0] for i in range(texts_stats.shape[0])]
texts_stats['date'] = [list(texts_stats.index)[i][1] for i in range(texts_stats.shape[0])]

texts_stats.columns = ['num of received sms','num of sent sms','num of received phone',
                       'num of sent phone','total len of received sms','total len of sent sms','mean len of received sms',
                      'mean len of sent sms','id','date']
new_texts = texts_stats
texts_stats = pd.DataFrame(columns = list(texts_stats.columns))
for i in texts_stats.columns:
  texts_stats[i] = list(new_texts[i])
del new_texts
texts_stats.head(10)
```

```python
final_output = pd.DataFrame()
patient_names = os.listdir(data_filepath)
for patient_name in patient_names:
  ## In this loop, we do the following:
  ## Get first and last date with the app on phone. These dates will be used as the time range in the subject's plots
  ## Some dates between the first and last date may not be in the summarized data frame because there were no incoming,
  ## outgoing, or missed calls on that date. We will add those dates with 0's for all the call/text statistics
  sub_calls = calls_stats.loc[list(calls_stats["id"] == patient_name),:]
  sub_texts = texts_stats.loc[list(texts_stats["id"] == patient_name),:]
  for j in range(2):
    if j==0:
      sub_data = sub_calls
    if j==1:
      sub_data = sub_texts
    if sub_data.shape[0] > 0:
      ##First date: take it from subject's identifiers file (first file if there are multiple)
      identifier_Files = os.listdir(data_filepath + "/" + patient_name +  "/identifiers")
      identifiers = pd.read_csv(data_filepath + "/" + patient_name +  "/identifiers/"+ identifier_Files[0], sep = ",")
      firstDate = identifiers["UTC time"]
      from_zone = tz.gettz('UTC')
      to_zone = tz.gettz('America/New_York')
      UTC = datetime.strptime(firstDate[0], '%Y-%m-%dT%H:%M:%S.%f') 
      EST = UTC.replace(tzinfo=from_zone).astimezone(to_zone)
      firstDate = EST.date()
      
      ##Last date: look at all the subject's directories (except survey) and find the latest datefor each directory
      directories = os.listdir(data_filepath + "/" + patient_name)
      directories = list(set(directories)-set(["survey_answers","survey_timings"]))
      lastDate = []
      
      for i in directories:
        files = os.listdir(data_filepath + "/" + patient_name + "/" + i)
        lastDate.append(files[-1])
      
      ##Converting dates to EST time zone
      from_zone = tz.gettz('UTC')
      to_zone = tz.gettz('America/New_York')
      UTC = [datetime.strptime(i, '%Y-%m-%d %H_%M_%S.csv') for i in lastDate]
      EST = [i.replace(tzinfo=from_zone).astimezone(to_zone) for i in UTC]
      date_list = [i.date() for i in EST]
      
      ##Get the latest date among all the directories
      lastDate = max(date_list)  
      
      ##Add zeroes for dates within the range but not in sub_data
      datesInRange = pd.date_range(start = firstDate, end = lastDate).tolist()
      complete_data = pd.DataFrame({"id":patient_name,"date":datesInRange})
      complete_data['date'] = complete_data['date'].astype(str)
      sub_data['date'] = sub_data['date'].astype(str)
      complete_data = complete_data.merge(sub_data,how='outer',on = ["id","date"]).fillna(0.0)
      
    if j==0:
      data1 = complete_data
    if j==1:
      data2 = complete_data
      data2 = data2.drop(["id","date"], axis=1)
      temp = pd.concat([data1,data2],axis=1)
      final_output = final_output.append(temp)
```
